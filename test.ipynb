```python
import numpy as np
import random
import csv
import mido
from mido import MidiFile
import os
```

```python
files = os.listdir("midi")
```

```python
for i in files:
    mid = MidiFile('midi/' + i)
    
    for j, track in enumerate(mid.tracks):
        print('Track {}: {}'.format(j, track.name), i)
        for msg in track:
            if msg.is_meta:
                print(msg)
        print(" ")
```
